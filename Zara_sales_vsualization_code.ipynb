```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load dataset
import pandas as pd

df = pd.read_csv("Zara_sales_data.csv", sep=';')

print(df.columns.tolist())

# Dataset Information
print(df.info())

# Display First 5 Rows
print(df.head())

# Statistical Summary
print(df.describe())

# Check Missing Values
print(df.isnull().sum())

# List All Columns
print(df.columns.tolist())
```

```python

# Sales Volume Distribution

plt.figure(figsize=(8,5))
plt.hist(df['Sales Volume'], bins=20)
plt.title('Sales Volume Distribution')
plt.xlabel('Sales Volume')
plt.ylabel('Frequency')
plt.show()
```

```python
import matplotlib.pyplot as plt

# Price Distribution
plt.figure(figsize=(8,5))
plt.hist(df['price'], bins=20)
plt.title('Price Distribution')
plt.xlabel('Price')
plt.ylabel('Frequency')
plt.show()
```

```python
# Product Position Analysis
df['Product Position'].value_counts().plot(kind='bar')
plt.title('Products by Position')
plt.xlabel('Position')
plt.ylabel('Count')
plt.show()
```

```python
# Promotion vs Sales Volume
sns.boxplot(x='Promotion', y='Sales Volume', data=df)
plt.title('Promotion vs Sales Volume')
plt.show()
```

```python
# Seasonal Product Analysis
sns.boxplot(
    x='Seasonal',
    y='Sales Volume',
    data=df
)

plt.title('Seasonal vs Sales Volume')
plt.show()
```

```python
# Category Wise Sales
df.groupby('Product Category')['Sales Volume'] \
    .sum() \
    .plot(kind='bar')

plt.title('Sales by Category')
plt.show()
```

```python
# Create estimated revenue
df['Revenue_Estimate'] = df['Sales Volume'] * df['price']

# Section Revenue Analysis
df.groupby('section')['Revenue_Estimate'] \
    .sum() \
    .plot(kind='bar')

plt.title('Revenue by Section')
plt.show()
```

```python
# Correlation Heatmap
sns.heatmap(df[['Sales Volume', 'price']].corr(),
            annot=True,
            cmap='Blues')
plt.show()
```

```python
# Price vs Sales Scatter Plot
sns.scatterplot(
    x='price',
    y='Sales Volume',
    hue='Promotion',
    data=df
)

plt.title('Price vs Sales')
plt.show()
```

```python
# Top Product Terms
df['terms'].value_counts().head(10) \
    .plot(kind='bar')

plt.title('Top Product Terms')
plt.show()
```
